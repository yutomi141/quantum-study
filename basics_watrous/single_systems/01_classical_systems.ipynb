# Unit 1: Classical Information
## 目的
量子情報科学の基礎となる「古典的確率論」を、線形代数で再構築する。
特に、確率ベクトルと確率行列（Stochastic Matrix）の性質をNumPyで実装し、後の量子状態（振幅）との対比の基準とする。
